```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Creo dos datasets de 50 muestras cada uno, con media y varianza distintas

```python
# Define parameters for dataset 1
mean1 = 10
std_dev1 = 2
num_samples = 50

# Generate dataset 1
data1 = np.random.normal(mean1, std_dev1, num_samples)

# Define parameters for dataset 2
mean2 = 15
std_dev2 = 3

# Generate dataset 2
data2 = np.random.normal(mean2, std_dev2, num_samples)
```

```python
# Create dataframes for each dataset
df1 = pd.DataFrame({'Value': data1, 'Dataset': 'Dataset 1'})
df2 = pd.DataFrame({'Value': data2, 'Dataset': 'Dataset 2'})

# Concatenate the dataframes
df = pd.concat([df1, df2])

# Reset index
df.reset_index(drop=True, inplace=True)

# Gr√°fica de los datos
plt.hist(data1, alpha=0.8, label="A")
plt.hist(data2, alpha=0.8, label="B")
plt.title("Distribuciones")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Generate labels for the datasets (assuming they are binary classification)
# For this example, let's randomly generate binary labels
labels1 = np.random.randint(2, size=len(data1))
labels2 = np.random.randint(2, size=len(data2))

# Compute ROC curve and AUC for dataset 1
fpr1, tpr1, thresholds1 = roc_curve(labels1, data1)
roc_auc1 = auc(fpr1, tpr1)

# Compute ROC curve and AUC for dataset 2
fpr2, tpr2, thresholds2 = roc_curve(labels2, data2)
roc_auc2 = auc(fpr2, tpr2)

# Plot ROC curves
plt.figure(figsize=(8, 6))

plt.plot(fpr1, tpr1, color='blue', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc1)
plt.plot(fpr2, tpr2, color='red', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc2)

plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

plt.show()
```
